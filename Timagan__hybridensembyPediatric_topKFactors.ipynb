````python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import zipfile
import re

def sanitize_filename(name):
    """Remplace tout caractÃ¨re non-alphanumÃ©rique par '_'."""
    return re.sub(r'[^A-Za-z0-9_-]+', '_', name)

def plot_bin_dist(cols, save_dir="distribution_attributs"):
    # CrÃ©er le dossier s'il n'existe pas
    os.makedirs(save_dir, exist_ok=True)

    for col in cols:
        if col in df.columns:
            plt.figure(figsize=(6,4))
            sns.countplot(x=col, data=df)
            plt.title(f'Distribution of {col}')
            plt.tight_layout()

            # Nettoyer le nom de fichier
            safe_col = sanitize_filename(col)
            filepath = os.path.join(save_dir, f"{safe_col}_distribution.png")

            # Sauvegarder avant d'afficher
            plt.savefig(filepath)
            plt.show()
            plt.close()

    # CrÃ©er un fichier ZIP contenant toutes les images
    zip_filename = f"{save_dir}.zip"
    with zipfile.ZipFile(zip_filename, "w", zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(save_dir):
            for file in files:
                zipf.write(os.path.join(root, file), arcname=file)

    print(f"Toutes les distributions ont Ã©tÃ© sauvegardÃ©es dans '{save_dir}/' et zippÃ©es dans '{zip_filename}'.")
````

````python
from sklearn.impute import SimpleImputer

# Normalisation des donnÃ©es
scaler = StandardScaler()

# Separate numeric and non-numeric columns for imputation
X_train_numeric = X_train.select_dtypes(include=np.number)
X_test_numeric = X_test.select_dtypes(include=np.number)

# Impute missing values with the mean (fitted only on numeric training data)
imputer = SimpleImputer(strategy='mean')
X_train_imputed_numeric = imputer.fit_transform(X_train_numeric)
X_test_imputed_numeric = imputer.transform(X_test_numeric)

# Convert imputed numeric arrays back to DataFrames to keep column names
X_train_imputed_numeric_df = pd.DataFrame(X_train_imputed_numeric, columns=X_train_numeric.columns, index=X_train.index)
X_test_imputed_numeric_df = pd.DataFrame(X_test_imputed_numeric, columns=X_test_numeric.columns, index=X_test.index)

# Combine imputed numeric features with original non-numeric features
# Need to ensure the column order matches X_train after one-hot encoding
# Let's get the non-numeric columns from the original X_train and X_test
X_train_non_numeric = X_train.select_dtypes(exclude=np.number)
X_test_non_numeric = X_test.select_dtypes(exclude=np.number)

# Combine the imputed numeric and original non-numeric columns.
# Need to be careful with column order to match the data used for model training.
# Assuming the column order of X_train after one-hot encoding and before scaling is the reference.
# Let's create a list of all columns from X_train (after get_dummies).
all_train_cols = X_train.columns

# Reconstruct the preprocessed training and test sets with imputed numeric and original non-numeric features
# Ensure the column order is consistent
X_train_processed = pd.concat([X_train_imputed_numeric_df, X_train_non_numeric], axis=1)
X_test_processed = pd.concat([X_test_imputed_numeric_df, X_test_non_numeric], axis=1)

# Reindex to ensure column order matches the original X_train columns after get_dummies
X_train_processed = X_train_processed.reindex(columns=all_train_cols)
X_test_processed = X_test_processed.reindex(columns=all_train_cols)

# Drop datetime columns before scaling
datetime_cols = ['Horodateur', 'Date of birth', 'Date of hospitallisation', 'Date of placement on oxygen', 'Date of discharge']
X_train_processed = X_train_processed.drop(columns=[col for col in datetime_cols if col in X_train_processed.columns])
X_test_processed = X_test_processed.drop(columns=[col for col in datetime_cols if col in X_test_processed.columns])


# Scale all features in the processed training and test sets
X_train_scaled = scaler.fit_transform(X_train_processed)
X_test_scaled = scaler.transform(X_test_processed)


# RÃ©Ã©chantillonnage SMOTE
# SMOTE is applied to X_train_scaled and Y_train needs to be a single column for SMOTE.
# Assuming the first target column is the one to be used for SMOTE.
y_train_res = Y_train['Discharge status'] # Select the 'Discharge status' target column for SMOTE

# Check class distribution
print("Class distribution before SMOTE:", y_train_res.value_counts())

# Check if the target variable has more than one class
if len(y_train_res.unique()) > 1:
    # Determine the minimum number of samples in any class
    min_samples = y_train_res.value_counts().min()

    # Set k_neighbors to be less than or equal to the minimum number of samples minus 1
    # Ensure k_neighbors is at least 1
    k_neighbors = max(1, min_samples - 1)

    sm = SMOTE(random_state=42, k_neighbors=k_neighbors)
    X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train_res)
    print("Class distribution after SMOTE:", y_train_res.value_counts())
else:
    print("SMOTE cannot be applied as the target variable has only one class.")
````

````python
# Function to evaluate models (modified for multi-class)
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    # For multi-class, precision, recall, and f1_score need 'average' parameter
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    # ROC AUC is typically for binary classification, so it's removed for multi-class

    cm = confusion_matrix(y_test, y_pred)

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'confusion_matrix': cm
    }
````

````python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def train_stacking_ensemble(models, X_train, y_train, X_test, y_test):
    """
    EntraÃ®ne un stacking ensemble de modÃ¨les calibrÃ©s avec gestion automatique des splits CV.
    """

    # VÃ©rifier la distribution de classes
    class_counts = pd.Series(y_train).value_counts()
    min_samples = class_counts.min()
    print(f"ðŸ“Š RÃ©partition des classes :\n{class_counts}\n")

    # Ajuster dynamiquement le nombre de splits
    n_splits_calibration = max(2, min(min_samples - 1, 5))
    print(f"ðŸ§© Utilisation de {n_splits_calibration} folds pour le stacking (min class = {min_samples})")

    # DÃ©finir les modÃ¨les de base et le mÃ©ta-modÃ¨le
    base_models = [(name.lower(), model) for name, model in models.items()]
    meta_model = LogisticRegression(max_iter=2000, class_weight='balanced', solver='lbfgs')

    # Construire le stacking ensemble
    stacking_model = StackingClassifier(
        estimators=base_models,
        final_estimator=meta_model,
        cv=n_splits_calibration,
        stack_method='predict_proba',
        n_jobs=-1
    )

    # EntraÃ®nement
    try:
        print("ðŸ”„ EntraÃ®nement du stacking ensemble...")
        stacking_model.fit(X_train, y_train)
        print("âœ… Stacking entraÃ®nÃ© avec succÃ¨s.")
    except ValueError as e:
        print(f"âš ï¸ Erreur lors de lâ€™entraÃ®nement : {e}")
        print("âž¡ï¸ Passage Ã  un entraÃ®nement sans CV (cv=2 forcÃ©).")
        stacking_model.cv = 2
        stacking_model.fit(X_train, y_train)

    # PrÃ©diction et Ã©valuation
    y_pred = stacking_model.predict(X_test)
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0),
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }

    print("\nðŸ“ˆ **Scores du stacking ensemble :**")
    for k, v in metrics.items():
        if k != 'confusion_matrix':
            print(f"  - {k.capitalize():<10}: {v:.4f}")

    # Matrice de confusion
    plt.figure(figsize=(8, 6))
    all_classes = np.unique(np.concatenate((y_test, y_pred)))
    sns.heatmap(metrics['confusion_matrix'], annot=True, fmt='d', cmap='Blues',
                xticklabels=all_classes, yticklabels=all_classes)
    plt.title("Matrice de Confusion - Stacking Ensemble")
    plt.xlabel("PrÃ©dit")
    plt.ylabel("RÃ©el")
    plt.tight_layout()
    plt.show()

    return stacking_model, metrics
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.calibration import CalibratedClassifierCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
import numpy as np

def train_and_evaluate_models(X_train, y_train, X_test, y_test):
    """
    EntraÃ®ne les modÃ¨les de base (LR, RF, XGB) + stacking, puis renvoie leurs mÃ©triques.
    """

    print("ðŸ”§ EntraÃ®nement et calibration des modÃ¨les de base...")

    # --- Calibration Fold dynamique ---
    min_samples_calibration = pd.Series(y_train).value_counts().min()
    n_splits_calibration = max(2, min_samples_calibration - 1)
    print(f"Calibration avec {n_splits_calibration} splits (min class = {min_samples_calibration})")

    # --- Logistic Regression ---
    lr = LogisticRegression(
        penalty='l1', solver='saga', max_iter=1000, class_weight='balanced'
    )
    lr_calibrated = CalibratedClassifierCV(lr, cv=n_splits_calibration, method='sigmoid')
    lr_calibrated.fit(X_train, y_train)

    # --- Random Forest ---
    rf = RandomForestClassifier(
        n_estimators=300, max_depth=10, class_weight='balanced', random_state=42
    )
    rf_calibrated = CalibratedClassifierCV(rf, cv=n_splits_calibration, method='sigmoid')
    rf_calibrated.fit(X_train, y_train)

    # --- XGBoost ---
    xgb = XGBClassifier(
        learning_rate=0.1,
        subsample=0.8,
        random_state=42,
        eval_metric='mlogloss',
        objective='multi:softprob',
        num_class=len(np.unique(y_train))
    )
    xgb_calibrated = CalibratedClassifierCV(xgb, cv=n_splits_calibration, method='sigmoid')
    xgb_calibrated.fit(X_train, y_train)

    print("âœ… ModÃ¨les calibrÃ©s.")

    # --- Stacking Ensemble ---
    print("ðŸ”„ EntraÃ®nement du stacking ensemble...")
    base_models = [
        ('lr', lr_calibrated),
        ('rf', rf_calibrated),
        ('xgb', xgb_calibrated)
    ]
    meta_model = LogisticRegression(class_weight='balanced')
    stacking_model = StackingClassifier(
        estimators=base_models,
        final_estimator=meta_model,
        cv=min(5, n_splits_calibration),
        stack_method='predict_proba',
        n_jobs=-1
    )
    stacking_model.fit(X_train, y_train)
    print("âœ… Stacking model entraÃ®nÃ©.")

    # --- Fonction dâ€™Ã©valuation ---
    def evaluate_model(model, X_test, y_test):
        y_pred = model.predict(X_test)
        return {
            'accuracy': accuracy_score(y_test, y_pred),
            'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
            'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
            'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0),
            'confusion_matrix': confusion_matrix(y_test, y_pred)
        }

    # --- Calcul des mÃ©triques ---
    lr_metrics = evaluate_model(lr_calibrated, X_test, y_test)
    rf_metrics = evaluate_model(rf_calibrated, X_test, y_test)
    xgb_metrics = evaluate_model(xgb_calibrated, X_test, y_test)
    stacking_metrics = evaluate_model(stacking_model, X_test, y_test)

    print("ðŸ“Š Ã‰valuation terminÃ©e.")

    return {
        'Logistic Regression': lr_metrics,
        'Random Forest': rf_metrics,
        'XGBoost': xgb_metrics,
        'Stacking Ensemble': stacking_metrics
    }
````

ðŸ‘‰ MÃªme quand certaines classes sont trop rares, on peut :

DÃ©sactiver la cross-validation interne du stacking (cv='prefit')

RÃ©utiliser les modÃ¨les dÃ©jÃ  calibrÃ©s (prÃ©-entraÃ®nÃ©s).

Combiner leurs prÃ©dictions directement â†’ stacking â€œmanuelâ€ (pondÃ©rÃ© ou par moyenne des probabilitÃ©s).

âœ… Voici une version corrigÃ©e et complÃ¨te de la fonction

Elle :

tente dâ€™utiliser StackingClassifier avec cv classique,

sinon bascule automatiquement vers une version stacking "prefit" basÃ©e sur les modÃ¨les calibrÃ©s dÃ©jÃ  entraÃ®nÃ©s.

````python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
import pandas as pd

def safe_stacking_training(models, X_train, y_train, X_test, y_test, n_splits_calibration=3):
    """
    Essaie d'entraÃ®ner un StackingClassifier. Si impossible (classes trop rares),
    bascule sur un stacking manuel utilisant les probabilitÃ©s moyennes.
    """

    print("ðŸ”„ EntraÃ®nement du stacking ensemble...")

    base_models = [
        ('lr', models['Logistic Regression']),
        ('rf', models['Random Forest']),
        ('xgb', models['XGBoost'])
    ]
    meta_model = LogisticRegression(class_weight='balanced', max_iter=1000)

    try:
        stacking_model = StackingClassifier(
            estimators=base_models,
            final_estimator=meta_model,
            cv=n_splits_calibration,
            stack_method='predict_proba',
            n_jobs=-1
        )
        stacking_model.fit(X_train, y_train)
        print("âœ… Stacking entraÃ®nÃ© avec succÃ¨s (classique).")

        y_pred = stacking_model.predict(X_test)

    except ValueError as e:
        print(f"âš ï¸ Stacking standard impossible : {e}")
        print("âž¡ï¸ Bascule vers un stacking manuel basÃ© sur les probabilitÃ©s moyennes.")

        # ProbabilitÃ©s combinÃ©es (moyenne pondÃ©rÃ©e)
        lr_proba = models['Logistic Regression'].predict_proba(X_test)
        rf_proba = models['Random Forest'].predict_proba(X_test)
        xgb_proba = models['XGBoost'].predict_proba(X_test)

        # PondÃ©ration Ã©gale (tu peux ajuster les poids ici)
        weights = np.array([1/3, 1/3, 1/3])
        avg_proba = (weights[0]*lr_proba + weights[1]*rf_proba + weights[2]*xgb_proba)

        y_pred = np.argmax(avg_proba, axis=1)
        stacking_model = None  # pas de vrai modÃ¨le sklearn

    # Ã‰valuation
    stacking_metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0),
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }

    return stacking_model, stacking_metrics
````

ðŸ§© Code Complet â€” Stacking PondÃ©rÃ© Adaptatif par Classe

Colle ce bloc Ã  la place du stacking pondÃ©rÃ© prÃ©cÃ©dent (ou juste aprÃ¨s lâ€™Ã©valuation de tes 3 modÃ¨les de base).

ðŸ§  Comment Ã§a fonctionne :

Pour chaque classe, on regarde quel modÃ¨le obtient le meilleur F1.

On calcule donc une matrice de poids (3 modÃ¨les Ã— n_classes).

Ces poids sont utilisÃ©s pour pondÃ©rer les probabilitÃ©s modÃ¨le par modÃ¨le :

Exemple : pour la classe 2, RF peut avoir 0.6, XGB 0.3, LR 0.1.

Chaque ligne du test est ensuite prÃ©dite Ã  partir des probabilitÃ©s combinÃ©es pondÃ©rÃ©es.

ðŸ“ˆ Avantages

âœ… Chaque modÃ¨le est rÃ©compensÃ© lÃ  oÃ¹ il est bon.
âœ… Plus robuste aux dÃ©sÃ©quilibres entre classes.
âœ… Aucune contrainte sur le nombre de splits (pas de CV error).

````python
import numpy as np
import pandas as pd
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)
import seaborn as sns
import matplotlib.pyplot as plt

# ===========================================================
# 1ï¸âƒ£ Ã‰valuation par modÃ¨le et par classe
# ===========================================================
def evaluate_model_per_class(model, X_test, y_test):
    """Retourne les F1 par classe + les scores globaux."""
    y_pred = model.predict(X_test)
    f1_per_class = f1_score(y_test, y_pred, average=None, zero_division=0)
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0),
        'f1_per_class': f1_per_class,
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }
    return metrics

# Ã‰valuation de chaque modÃ¨le
lr_metrics = evaluate_model_per_class(lr_calibrated, X_test, y_test)
rf_metrics = evaluate_model_per_class(rf_calibrated, X_test, y_test)
xgb_metrics = evaluate_model_per_class(xgb_calibrated, X_test, y_test)

# Classes prÃ©sentes
classes = np.unique(y_test)
n_classes = len(classes)

# ===========================================================
# 2ï¸âƒ£ PondÃ©ration adaptative par classe
# ===========================================================
# F1 par modÃ¨le et par classe
f1_matrix = np.vstack([
    lr_metrics['f1_per_class'],
    rf_metrics['f1_per_class'],
    xgb_metrics['f1_per_class']
])

# Normalisation par colonne (chaque classe)
weights_per_class = f1_matrix / f1_matrix.sum(axis=0, keepdims=True)

weights_df = pd.DataFrame(
    weights_per_class.T,
    columns=['LR', 'RF', 'XGB'],
    index=[f'Classe {c}' for c in classes]
)

print("ðŸ“Š PondÃ©rations dynamiques par classe :")
display(weights_df.round(3))

# ===========================================================
# 3ï¸âƒ£ PrÃ©dictions du stacking adaptatif
# ===========================================================
# ProbabilitÃ©s de chaque modÃ¨le
lr_proba = lr_calibrated.predict_proba(X_test)
rf_proba = rf_calibrated.predict_proba(X_test)
xgb_proba = xgb_calibrated.predict_proba(X_test)

# PondÃ©ration par classe (diffÃ©rente pour chaque colonne)
stacking_proba = np.zeros_like(lr_proba)
for j, c in enumerate(classes):
    stacking_proba[:, j] = (
        weights_per_class[0, j] * lr_proba[:, j] +
        weights_per_class[1, j] * rf_proba[:, j] +
        weights_per_class[2, j] * xgb_proba[:, j]
    )

# Classe finale = argmax
stacking_preds = np.argmax(stacking_proba, axis=1)

# ===========================================================
# 4ï¸âƒ£ Ã‰valuation finale du stacking adaptatif
# ===========================================================
stacking_metrics = {
    'accuracy': accuracy_score(y_test, stacking_preds),
    'precision': precision_score(y_test, stacking_preds, average='weighted', zero_division=0),
    'recall': recall_score(y_test, stacking_preds, average='weighted', zero_division=0),
    'f1': f1_score(y_test, stacking_preds, average='weighted', zero_division=0),
    'confusion_matrix': confusion_matrix(y_test, stacking_preds)
}

print("\nâœ… Stacking pondÃ©rÃ© adaptatif terminÃ©.")

# ===========================================================
# 5ï¸âƒ£ Comparaison visuelle
# ===========================================================
metrics_df = pd.DataFrame({
    'Logistic Regression': lr_metrics,
    'Random Forest': rf_metrics,
    'XGBoost': xgb_metrics,
    'Stacking Adaptatif': stacking_metrics
}).drop('confusion_matrix').drop('f1_per_class', errors='ignore')

plt.figure(figsize=(12, 8))
metrics_df.T.plot(kind='bar', colormap='viridis')
plt.title('Comparaison des performances (Stacking Adaptatif par Classe)')
plt.ylabel('Score')
plt.xticks(rotation=15)
plt.ylim(0.7, 1.0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# ===========================================================
# 6ï¸âƒ£ Matrice de confusion + rapport
# ===========================================================
plt.figure(figsize=(8, 6))
sns.heatmap(stacking_metrics['confusion_matrix'], annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.xlabel('PrÃ©dit')
plt.ylabel('RÃ©el')
plt.title('Matrice de Confusion - Stacking Adaptatif')
plt.tight_layout()
plt.show()

print("\nClassification Report - Stacking Adaptatif :")
print(classification_report(y_test, stacking_preds, zero_division=0))
````

````python
# ===============================================================
# 6ï¸âƒ£ Visualisation : Poids combinÃ©s (classe Ã— contexte clinique)
# ===============================================================
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# --- Fonction pour extraire les poids contextuels utilisÃ©s ---
def dynamic_contextual_weights(features):
    """
    Recalcule les pondÃ©rations dynamiques selon les signes cliniques dâ€™un patient.
    UtilisÃ© uniquement pour la visualisation (doit correspondre Ã  ta logique rÃ©elle).
    """
    features = pd.Series(features)

    # Exemple de logique clinique simplifiÃ©e
    if features.get('Fever', 0) == 1 or features.get('Signs of Respiratory Distress', 0) == 1:
        return {'lr': 0.6, 'rf': 0.2, 'xgb': 0.2}
    elif pd.notna(features.get('Age (days)', np.nan)) and features.get('Age (days)') < (24 * 30.44):
        return {'lr': 0.2, 'rf': 0.3, 'xgb': 0.5}
    else:
        return {'lr': 0.4, 'rf': 0.4, 'xgb': 0.2}


# --- Fonction pour extraire les poids de performance des modÃ¨les (par classe) ---
def compute_class_weights(lr_model, rf_model, xgb_model, X_test, y_test):
    """
    Calcule les poids de performance (par classe) des modÃ¨les sur lâ€™ensemble de test.
    Ces poids sont utilisÃ©s pour combiner les prÃ©dictions du stacking pondÃ©rÃ©.
    """
    models = [lr_model, rf_model, xgb_model]
    class_weights = []

    for model in models:
        # ProbabilitÃ©s prÃ©dictives
        proba = model.predict_proba(X_test)
        preds = np.argmax(proba, axis=1)

        # Calcul F1-score par classe
        from sklearn.metrics import f1_score
        unique_classes = np.unique(y_test)
        f1_scores = f1_score(y_test, preds, average=None, labels=unique_classes)
        # Normalisation
        f1_scores /= np.sum(f1_scores)
        class_weights.append(f1_scores)

    return np.array(class_weights)


# --- Fonction principale de visualisation ---
def visualize_hybrid_weights(X_test, y_test, class_weights, sample_indices=None):
    """
    Visualise les pondÃ©rations combinÃ©es (clinique Ã— performance)
    pour plusieurs patients, et gÃ©nÃ¨re une carte de chaleur globale.
    """
    if sample_indices is None:
        # SÃ©lectionne 5 patients au hasard pour illustrer
        sample_indices = np.random.choice(len(X_test), size=min(5, len(X_test)), replace=False)

    records = []
    for i in sample_indices:
        patient = X_test.iloc[i]
        clinical_w = dynamic_contextual_weights(patient)

        for j, model in enumerate(['LR', 'RF', 'XGB']):
            for cls in range(class_weights.shape[1]):
                records.append({
                    'Patient': f'P{i+1}',
                    'Model': model,
                    'Class': f'C{cls}',
                    'Clinical Weight': clinical_w[model.lower()],
                    'Class Weight': class_weights[j, cls],
                    'Combined Weight': clinical_w[model.lower()] * class_weights[j, cls],
                    'Age (days)': patient.get('Age (days)', np.nan),
                    'Fever': patient.get('Fever', np.nan)
                })

    df_weights = pd.DataFrame(records)

    # --- Graphique 1 : Poids combinÃ©s par modÃ¨le et classe ---
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_weights, x='Class', y='Combined Weight', hue='Model')
    plt.title("Poids combinÃ©s (Classe Ã— ModÃ¨le) pour quelques patients")
    plt.ylabel("Poids combinÃ© (clinique Ã— performance)")
    plt.xlabel("Classe prÃ©dite")
    plt.legend(title="ModÃ¨le de base")
    plt.tight_layout()
    plt.show()

    # --- Graphique 2 : Influence clinique (Ã¢ge + fiÃ¨vre) ---
    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        data=df_weights,
        x='Age (days)', y='Combined Weight',
        hue='Model', size='Fever',
        sizes=(50, 150), alpha=0.8
    )
    plt.title("Impact de lâ€™Ã¢ge et de la fiÃ¨vre sur les pondÃ©rations des modÃ¨les")
    plt.tight_layout()
    plt.show()

    # --- Graphique 3 : Carte de chaleur globale des poids combinÃ©s ---
    heatmap_data = (
        df_weights.groupby(['Model', 'Class'])['Combined Weight']
        .mean()
        .unstack()
        .fillna(0)
    )

    plt.figure(figsize=(8, 5))
    sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt=".2f")
    plt.title("Moyenne des poids combinÃ©s par modÃ¨le et par classe")
    plt.xlabel("Classe")
    plt.ylabel("ModÃ¨le de base")
    plt.tight_layout()
    plt.show()

    return df_weights


# ===============================================================
# 7ï¸âƒ£ Appel final de visualisation aprÃ¨s lâ€™entraÃ®nement
# ===============================================================
class_weights = compute_class_weights(lr_calibrated, rf_calibrated, xgb_calibrated, X_test, y_test)
weights_df = visualize_hybrid_weights(X_test, y_test, class_weights)
````

````python
import numpy as np
import pandas as pd
import joblib

class ManualWeightedStacker:

    def __init__(self, base_models, target_name):
        """
        base_models = dict {
            'XGBoost': trained_model,
            'RandomForest': trained_model,
            'LogisticRegression': trained_model
        }
        """
        self.base_models = base_models
        self.target_name = target_name
        self.class_weights = None
        self.model_names = list(base_models.keys())

    # ----------------------------------------------------------------------
    # 1) CALCUL DES POIDS DE PERFORMANCE (VALIDATION)
    # ----------------------------------------------------------------------
    def fit_validation_weights(self, X_val, y_val):
        """
        Calcule les F1 par classe pour chaque modÃ¨le.
        Normalise pour obtenir des poids (somme = 1).
        """
        from sklearn.metrics import f1_score

        n_models = len(self.model_names)
        classes = np.unique(y_val)
        C = len(classes)

        weights = np.zeros((n_models, C))

        for i, name in enumerate(self.model_names):
            model = self.base_models[name]
            proba = model.predict_proba(X_val)
            preds = np.argmax(proba, axis=1)

            f1 = f1_score(y_val, preds, average=None, labels=classes, zero_division=0)

            # Normalisation sÃ©curisÃ©e
            if f1.sum() == 0:
                weights[i] = np.ones(C) / C
            else:
                weights[i] = f1 / f1.sum()

        self.class_weights = weights
        return weights



    def _clinical_weights(self, row):
        """
        PondÃ©ration clinique avancÃ©e pour bronchiolite pÃ©diatrique.
        OptimisÃ©e pour XGBoost / RF / LogReg.
        """

        # -----------------------------------------------------------
        # 1) Variables utiles (sÃ©curisÃ©es)
        # -----------------------------------------------------------
        age_days = row.get('Age_(days)', np.nan)
        fever = row.get('Fever', 0)
        distress = row.get('Signs_of_Respiratory_Distress', 0)
        nasal_flaring = row.get('Nasal_flaring', 0)
        wheezes = row.get('Wheezes', 0)
        chest_crackles = row.get('Lungs_crackles', 0)
        saturation = row.get('Oxygen_saturation_90', np.nan)  # fallback si multiple colonnes

        # comorbiditÃ©s
        CHD = row.get('Congenital_Heart_Disease', 0)
        HIV = row.get('HIV__AIDS', 0)
        malnutrition = row.get('Malnutrition', 0)

        # traitements reÃ§us
        bronchodilator = row.get('Bronchodilators', 0)
        oxygen_admin = row.get('Oxygen_during_hospitalisation(_excl.admission)', 0)
        antibiotics = row.get('Antibiotics_', 0)

        # -----------------------------------------------------------
        # 2) Scores cliniques (logique mÃ©dicale alignÃ©e CHU pÃ©diatrique)
        # -----------------------------------------------------------
        severity_score = 0

        # A) Signes vitaux dangereux
        if saturation < 90:
            severity_score += 3
        elif saturation < 94:
            severity_score += 2

        if fever >= 38:
            severity_score += 1
        if fever >= 39:
            severity_score += 2

        if distress == 1:
            severity_score += 3
        if nasal_flaring == 1:
            severity_score += 2
        if wheezes == 1:
            severity_score += 1
        if chest_crackles == 1:
            severity_score += 1

        # B) ComorbiditÃ©s majorantes (guidelines OMS)
        if CHD == 1:
            severity_score += 3
        if HIV == 1:
            severity_score += 2
        if malnutrition == 1:
            severity_score += 2

        # C) Nouveau-nÃ© et <3 mois = haut risque
        if age_days < 30:
            severity_score += 3
        elif age_days < 90:
            severity_score += 2

        # D) DÃ©jÃ  sous oxygÃ¨ne â†’ cas sÃ©vÃ¨re
        if oxygen_admin == 1:
            severity_score += 2

        # -----------------------------------------------------------
        # 3) PondÃ©ration finale selon la sÃ©vÃ©ritÃ©
        # -----------------------------------------------------------
        if severity_score >= 8:   # Cas sÃ©vÃ¨re
            return {
                'XGBoost': 0.65,
                'RandomForest': 0.25,
                'LogisticRegression': 0.10
            }

        elif severity_score >= 4:  # Cas modÃ©rÃ©
            return {
                'XGBoost': 0.45,
                'RandomForest': 0.35,
                'LogisticRegression': 0.20
            }

        else:  # Cas lÃ©ger / stable
            return {
                'XGBoost': 0.30,
                'RandomForest': 0.40,
                'LogisticRegression': 0.30
            }

    # ----------------------------------------------------------------------
    # 2) CALCUL DES POIDS CLINIQUES CONTEXTUELS (PAR PATIENT)
    # ----------------------------------------------------------------------
    def _clinical_context(self, row):
        """
        RÃ¨gles EXACTES adaptÃ©es Ã  tes colonnes.
        """
        # Cas sÃ©vÃ¨re respiratoire
        if row.get('Fever', 0) == 1 or row.get('Signs_of_Respiratory_Distress', 0) == 1:
            return {'XGBoost': 0.6, 'RandomForest': 0.2, 'LogisticRegression': 0.2}

        # Nouveau-nÃ© (< 1 mois)
        if row.get('Age_(days)', 99999) < 30:
            return {'XGBoost': 0.5, 'RandomForest': 0.3, 'LogisticRegression': 0.2}

        # Default
        return {'XGBoost': 0.4, 'RandomForest': 0.4, 'LogisticRegression': 0.2}


    # ----------------------------------------------------------------------
    # 3) PREDICTION COMBINÃ‰E
    # ----------------------------------------------------------------------
    def predict(self, X_df):
        if self.class_weights is None:
            raise ValueError("You must call fit_validation_weights() first.")

        X_df = pd.DataFrame(X_df, columns=X_df.columns)  # ensure DF

        n_samples = len(X_df)
        any_model = next(iter(self.base_models.values()))
        C = any_model.predict_proba(X_df).shape[1]

        final_proba = np.zeros((n_samples, C))

        for i in range(n_samples):
            row = X_df.iloc[i]
            clinical_w = self._clinical_weights(row)

            combined = np.zeros(C)

            for m_idx, name in enumerate(self.model_names):
                model = self.base_models[name]
                p = model.predict_proba([row.values])[0]

                combined += p * self.class_weights[m_idx] * clinical_w[name]

            if combined.sum() > 0:
                combined /= combined.sum()

            final_proba[i] = combined

        return np.argmax(final_proba, axis=1), final_proba

    # ----------------------------------------------------------------------
    # 4) SAUVEGARDE & CHARGEMENT
    # ----------------------------------------------------------------------
    def save(self, path):
        joblib.dump({
            'base_models': self.base_models,
            'class_weights': self.class_weights,
            'target_name': self.target_name
        }, path)

    @staticmethod
    def load(path):
        data = joblib.load(path)
        obj = ManualWeightedStacker(data['base_models'], data['target_name'])
        obj.class_weights = data['class_weights']
        return obj
````

# âœ… Top-K stacking

âœ… 1. Fonction : calcul des importances stacking

ðŸ‘‰ Ã€ coller dans ton notebook (ou un fichier stacking_importance.py).

````python
import numpy as np
import pandas as pd

def compute_stacking_feature_importance(
    base_models: dict,
    model_weights: dict,
    feature_names: list,
    topk=20
):
    """
    Calcule les importances du stacking comme:
        importance = somme (poids_model Ã— importance_model)
    base_models : dict comme {'LR': model, 'RF': model, 'XGB': model}
    model_weights : dict comme {'LR':0.3,'RF':0.4,'XGB':0.3}
    feature_names : X_test_imp.columns
    """

    stacking_importance = pd.Series(0.0, index=feature_names)

    for name, model in base_models.items():
        w = model_weights[name]

        # 1) rÃ©cupÃ©rer importance du modÃ¨le
        if hasattr(model, "feature_importances_"):
            imp = model.feature_importances_
        elif hasattr(model, "coef_"):
            imp = np.mean(np.abs(model.coef_), axis=0)
        else:
            continue

        # aligner sur nom des colonnes
        imp_series = pd.Series(imp, index=feature_names)

        # 2) ajouter pondÃ©rÃ©
        stacking_importance += w * imp_series

    # trier
    stacking_importance = stacking_importance.sort_values(ascending=False)

    # top K
    return stacking_importance.head(topk)
````

````python
def compute_local_stacking_contributions(stacker, base_models, model_weights, patient_row):
    """
    Calcule contribution locale par feature pour un patient.
    = somme (poids_model Ã— grad approx : importance_model Ã— valeur_normalisÃ©e)

    patient_row : Series contenant 1 ligne du X_test_imp
    """
    contributions = pd.Series(0.0, index=patient_row.index)

    # normalisation simple (evite influence brute des Ã©chelles)
    norm_row = (patient_row - patient_row.mean()) / (patient_row.std() + 1e-6)

    for name, model in base_models.items():
        w = model_weights[name]

        # importance du modÃ¨le
        if hasattr(model, "feature_importances_"):
            imp = pd.Series(model.feature_importances_, index=patient_row.index)
        elif hasattr(model, "coef_"):
            imp = pd.Series(np.mean(np.abs(model.coef_), axis=0), index=patient_row.index)
        else:
            continue

        # contribution = poids_model Ã— importance Ã— valeur_normÃ©e
        contributions += w * imp * norm_row

    return contributions.sort_values(ascending=False)
````

````python
# stacking_explainability.py
"""
Module pour manuel stacking pondÃ©rÃ© + explainability locale/global + export PDF/CSV.
Classe principale: ManualWeightedStacker
Fonctions utilitaires: compute_local_contributions, plot_local_contributions, export_patient_report_pdf
"""

import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fpdf import FPDF  # pip install fpdf
from typing import Dict, List, Optional

sns.set_style("whitegrid")


class ManualWeightedStacker:
    def __init__(self, base_models: Dict[str, object], model_names: List[str] = None):
        """
        base_models : dict like { 'XGBoost': model_obj, 'RandomForest': model_obj, ... }
        model_names : optional ordering list of names (should match keys)
        """
        self.base_models = base_models
        self.model_names = model_names or list(base_models.keys())
        self.class_weights_by_model = {}  # per-class performance weights
        self.global_model_weights = {name: 1.0 / len(self.model_names) for name in self.model_names}
        self.is_fitted = True  # models must be pre-trained

    def fit_validation_weights(self, X_val: pd.DataFrame, y_val: pd.Series, metric: str = "f1", average="weighted"):
        """
        Compute model weights from validation performance per class.
        Stores:
          - self.class_weights_by_model[name] = f1 per class normalized
          - self.global_model_weights[name] = overall f1 normalized
        """
        from sklearn.metrics import f1_score

        overall_scores = {}
        class_weights = {}

        for name in self.model_names:
            model = self.base_models[name]
            if hasattr(model, "predict"):
                y_pred = model.predict(X_val)
            else:
                raise ValueError(f"Model {name} missing predict()")

            # overall score
            try:
                overall = f1_score(y_val, y_pred, average=average, zero_division=0)
            except Exception:
                overall = 0.0
            overall_scores[name] = overall

            # per-class f1 (returns array aligned to unique labels)
            labels = np.unique(np.concatenate([np.array(y_val), np.array(y_pred)]))
            per_class = []
            for lab in labels:
                per_class.append(f1_score(y_val, y_pred, labels=[lab], average="binary", zero_division=0))
            per_class = np.array(per_class, dtype=float)

            # normalize per-class to sum=1 (avoid all-zero)
            if per_class.sum() == 0:
                per_class_norm = np.ones_like(per_class) / len(per_class)
            else:
                per_class_norm = per_class / per_class.sum()

            class_weights[name] = {
                "labels": labels,
                "per_class": per_class_norm
            }

        # normalize overall to sum=1
        vals = np.array(list(overall_scores.values()), dtype=float)
        if vals.sum() == 0:
            norm_vals = np.ones_like(vals) / len(vals)
        else:
            norm_vals = vals / vals.sum()

        self.global_model_weights = {name: w for name, w in zip(self.model_names, norm_vals)}
        self.class_weights_by_model = class_weights
        return self.global_model_weights, self.class_weights_by_model

    def predict(self, X: pd.DataFrame, dynamic_weights_fn=None, return_proba: bool = False):
        """
        Predict by combining base model probabilities using:
          combined_proba = sum( model_weight * model.predict_proba(X) )
        dynamic_weights_fn: optional function(patient_row: pd.Series) -> dict model_name->weight (sums to 1)
        Returns predictions (and proba if requested).
        """
        # Ensure X is DataFrame
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        # accumulate weighted probabilities
        # Get classes ordering from first model that supports predict_proba
        proba_acc = None
        classes = None

        for name in self.model_names:
            model = self.base_models[name]
            # get probs
            if hasattr(model, "predict_proba"):
                p = model.predict_proba(X)
                # model-specific class order
                model_classes = getattr(model, "classes_", None)
            else:
                # fallback: use one-hot from predict results (not ideal)
                preds = model.predict(X)
                uniq = np.unique(preds)
                p = np.zeros((len(preds), len(uniq)))
                for i, v in enumerate(preds):
                    p[i, np.where(uniq == v)[0][0]] = 1.0
                model_classes = uniq

            if proba_acc is None:
                # initialize accumulator with zeros, and set classes reference
                proba_acc = np.zeros_like(p, dtype=float)
                classes = model_classes

            # alignment: convert this model's p to the accumulator columns (classes)
            if not np.array_equal(model_classes, classes):
                # build mapping
                new_p = np.zeros((p.shape[0], len(classes)))
                for idx, c in enumerate(classes):
                    if c in model_classes:
                        col = list(model_classes).index(c)
                        new_p[:, idx] = p[:, col]
                    else:
                        new_p[:, idx] = 0.0
                p = new_p

            # compute weight: either dynamic per-row or static global
            if dynamic_weights_fn is None:
                w = self.global_model_weights.get(name, 1.0 / len(self.model_names))
                proba_acc += w * p
            else:
                # dynamic weights per-row: aggregate across rows
                dyn_weights_matrix = np.vstack([np.array([dynamic_weights_fn(row).get(name, 0.0)
                                                         for name in self.model_names]) for _, row in X.iterrows()])
                # dyn_weights_matrix shape = (n_rows, n_models); pick this model column
                model_col = self.model_names.index(name)
                col_weights = dyn_weights_matrix[:, model_col].reshape(-1, 1)
                proba_acc += col_weights * p

        # final preds
        preds = classes[np.argmax(proba_acc, axis=1)]
        if return_proba:
            return preds, proba_acc, classes
        return preds

    def compute_local_contributions(self, patient_row: pd.Series, model_weights: Optional[Dict[str, float]] = None):
        """
        Returns a Series indexed by feature with contribution scores for one patient.
        Contribution per model = model_weight * model_importance * normalized_value.
        model_weights overrides self.global_model_weights (if provided).
        """
        if not isinstance(patient_row, pd.Series):
            patient_row = pd.Series(patient_row)

        weights = model_weights if model_weights is not None else self.global_model_weights
        # normalize patient features (robust)
        vals = patient_row.astype(float)
        vals_norm = (vals - vals.mean()) / (vals.std() + 1e-9)

        contrib = pd.Series(0.0, index=patient_row.index, dtype=float)

        for name in self.model_names:
            model = self.base_models[name]
            w = weights.get(name, 0.0)

            # get importances
            if hasattr(model, "feature_importances_"):
                imp = pd.Series(model.feature_importances_, index=patient_row.index)
            elif hasattr(model, "coef_"):
                coef = model.coef_
                if coef.ndim > 1:
                    imp = pd.Series(np.mean(np.abs(coef), axis=0), index=patient_row.index)
                else:
                    imp = pd.Series(np.abs(coef), index=patient_row.index)
            else:
                # skip models w/o feature-level importances
                continue

            # safe align (some columns may differ) -> reindex
            imp = imp.reindex(patient_row.index).fillna(0.0)

            contrib += w * imp * vals_norm

        # absolute and sort descending
        contrib = contrib.abs().sort_values(ascending=False)
        return contrib

    def save(self, path: str):
        """Save the ManualWeightedStacker instance to disk (joblib)."""
        joblib.dump(self, path)
        return path


# -----------------------
# Utilitaires externes
# -----------------------
def plot_local_contributions(local_series: pd.Series, topk: int = 10, patient_id: Optional[str] = None, figsize=(8, 5)):
    top = local_series.head(topk)
    plt.figure(figsize=figsize)
    sns.barplot(x=top.values, y=top.index)
    title = f"Top-{topk} factors"
    if patient_id is not None:
        title += f" - Patient {patient_id}"
    plt.title(title)
    plt.tight_layout()
    plt.show()


def export_patient_report_pdf(output_path: str,
                              patient_series: pd.Series,
                              predicted_class,
                              local_contrib: pd.Series,
                              topk: int = 8,
                              extra_meta: dict = None):
    """
    Create a simple PDF report summarizing prediction + topk factors.
    Uses fpdf (pip install fpdf).
    """
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    pdf.cell(0, 8, "Clinical Prediction Report", ln=1, align="C")
    pdf.ln(4)

    # metadata
    if extra_meta:
        for k, v in extra_meta.items():
            pdf.cell(0, 6, f"{k}: {v}", ln=1)

    pdf.ln(4)
    pdf.cell(0, 6, f"Predicted class: {predicted_class}", ln=1)
    pdf.ln(4)

    pdf.cell(0, 6, "Top contributing features:", ln=1)
    pdf.ln(2)

    top = local_contrib.head(topk)
    for feat, val in top.items():
        pdf.cell(0, 6, f"{feat} : {val:.4f}", ln=1)

    pdf.ln(8)
    pdf.cell(0, 6, "Feature values:", ln=1)
    pdf.ln(2)
    # show some patient features
    for feat in patient_series.index[:40]:  # avoid huge pdf
        try:
            pdf.cell(0, 5, f"{feat}: {patient_series[feat]}", ln=1)
        except Exception:
            pass

    pdf.output(output_path)
    return output_path
````
